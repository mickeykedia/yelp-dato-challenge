### Number of Friends 


```python
# constructing the social network of users
edges = gl.load_sframe('yelp_dataset_challenge_academic_dataset/edges/')
```

```python
uf_graph = gl.SGraph(users, edges, 'user_id', 'user_id', 'friend_id')
```

```python
dc = gl.degree_counting.create(uf_graph)
```

```python
vertices = dc.graph.get_vertices()
```

```python
vertices.head(1)
```

```python
# Number of vertices where in-degreee is not equal to out-degree
len(vertices[vertices['in_degree']!=vertices['out_degree']])
```

```python
vertices['in_degree'].sketch_summary()
```

### Fan distribution

```python
users['fans'].sketch_summary()
```

```python
# joining with the vertices to look for fans of users with more friends  
users = vertices.join(users, on={'__id':'user_id'})
```

```python
users.head(1)
```

```python
users[users['in_degree']>7]['fans'].sketch_summary()
```
